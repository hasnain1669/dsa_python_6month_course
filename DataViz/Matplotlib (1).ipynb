## Visualization using matplotlib and pandas

##### The aim of this notebook is to show the visualization of using matplotlib and pandas

Showing the Line plots, bar plots, hist plots, pie plots, 2D plots and 3D plots.

```python
# import the library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('default')
```

## Subplots

```python
# A different way to plot graphs
batter.head()
```

```python
# 1st way
plt.figure(figsize=(12, 6))

plt.scatter(batter['avg'], batter['strike_rate'])
plt.title('Average vs Strike_rate')
plt.xlabel('Average')
plt.ylabel('Strike_rate')

plt.show()
```

```python
# 2nd way
fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(batter['avg'], batter['strike_rate'], color='red', marker='+')
ax.set_title('Average vs Strike_rate')
ax.set_xlabel('Average')
ax.set_ylabel('Strike_Rate')

fig.show()
```

```python
# (fig) 1st object and (ax) is the axis object
fig, ax = plt.subplots()

# passing the values of (axis) in ax
ax.plot(batter['avg'], batter['strike_rate'], marker='o')
ax.set_title('Average vs Strike_rate')
ax.set_xlabel('Average')
ax.set_ylabel('Strike_Rate')
```

```python
# subplots
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(12, 6))

ax[0].scatter(batter['avg'], batter['strike_rate'], color='red')
ax[1].scatter(batter['avg'], batter['runs'])

ax[0].set_title('Avg vs Strike_Rate')
ax[0].set_ylabel('Strike_Rate')

ax[1].set_title('Avg vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
# subplots scatter and histogram
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

ax[0, 0].scatter(batter['avg'], batter['strike_rate'], color='red')
ax[0, 1].scatter(batter['avg'], batter['runs'])
ax[1, 0].hist(batter['avg'])
ax[1, 1].hist(batter['runs'])
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'], batter['strike_rate'], color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batter['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['avg'])
```

## 3D Plots

```python
# scatter plot
fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(batter['runs'], batter['avg'], batter['strike_rate'])
ax.set_title('IPL batsman analysis')

ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('SR')
```

```python
# line plot
x = [0, 1, 5, 13]
y = [0, 10, 13, 8]
z = [0, 13, 20, 5]

fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(x, y, z, s=[100, 100, 100, 100])
ax.plot3D(x, y, z, color='red')
```

```python
# 3D Surface plots
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x, y)
xx.shape, yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12, 8))

ax = plt.subplot(projection='3d')

P3d = ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(P3d)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12, 8))

ax = plt.subplot(projection='3d')

P3d = ax.plot_surface(xx, yy, z, cmap='viridis')

fig.colorbar(P3d)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12, 8))

ax = plt.subplot()

P3d = ax.contourf(xx, yy, z, cmap='viridis')

fig.colorbar(P3d)
```

```python
z = np.sin(xx) + np.log1p(yy)

fig = plt.figure(figsize=(12, 8))

ax = plt.subplot(projection='3d')

P3d = ax.plot_surface(xx, yy, z, cmap='viridis')

fig.colorbar(P3d)
```

```python
# contour plot
fig = plt.figure(figsize=(12, 8))

ax = plt.subplot(projection='3d')

P3d = ax.contour3D(xx, yy, z, cmap='viridis')
fig.colorbar(P3d)
```

```python
# contour plot
fig = plt.figure(figsize=(12, 8))

ax = plt.subplot()

P3d = ax.contourf(xx, yy, z, cmap='viridis')
fig.colorbar(P3d)
```

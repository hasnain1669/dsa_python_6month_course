1. Relational Plot

To see the statistical relation between 2 or more variables.

i. Bivariate Analysis

ii. Plots under this section

- Scatterplot
- Lineplot
- Relplot

```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px # use for datasets
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axes ;eve; function -> rectangle shape
sns.scatterplot(data=tips, x='total_bill', y='tip', size='size')
```

```python
# scatter plot -> figure level function -> square shape
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time')
```

```python
# import the dataset gapminder
gap = px.data.gapminder()
gap
```

```python
pak_df = gap[gap['country'] == 'Pakistan']
pak_df
```

```python
# line plot -> axes level function
sns.lineplot(data=pak_df, x='year', y='lifeExp')
```

```python
# using relplot -> figure level function
sns.relplot(data=pak_df, x='year', y='lifeExp', kind='line')
```

```python
pic = gap[gap['country'].isin(['India', 'Pakistan', 'China'])]
pic
```

```python
sns.relplot(kind='line', data=pic, x='year', y='lifeExp', hue='country', )
```

```python
# Continent
bip = gap[gap['country'].isin(['Germany', 'Pakistan', 'Brazil'])]
sns.relplot(kind='line', data=bip, x='year', y='lifeExp', hue='country', style='continent', size='continent')
```

```python
# Facet plot -> It will work with figure level function -> it means it work with relplot -> it won't work on axes level function
sns.relplot(data=tips, x='total_bill', y='tip', col='sex', row='smoker', kind='scatter') # row='var', col='var', kind='line'
```

```python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=4)
```

2. Distribution Plots

- Used for univariate analysis
- Used to find out the distribution
- Range of the observation
- Central Tendency
- Is the data bimodal?
- Are there outliers?
- Plots under distribution plot

    i- histplot

    ii- kdeplot
    
    iii- rugplot
    
    iv- displot
    
    v- ecdfplot

```python
# ploting univariate  histogram
# figure level function -> displot
# axes level function -> histplot, kdeplot, rugplot
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# bins parameter
sns.displot(data=tips, x='total_bill', kind='hist', bins=20, hue='sex')
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# countplot (histogram acting like a countplot)
sns.displot(data=tips, x='day', kind='hist')
```

```python
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
# facet plot, row, col, this won't work on histplot function, because histplot is axes level function
sns.displot(data=titanic, x='age', kind='hist', col='sex', element='step')
```

```python
# KDE plot (kernel density estimation)
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate

sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex')
```

```python
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True, height=4, aspect=2)
```

```python
# rugplot
# plot marginal distributions by drawing ticks along the x and y axes,
# this function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip', hue='sex')
sns.displot(data=tips, x='total_bill', y='tip', kind='hist', hue='sex')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

sns.kdeplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
# ecdfplot

sns.ecdfplot(data=titanic, x='age', hue='sex')
```

5. Matrix Plot
- Heatmap
- Clustermap

```python
# heatmap
# plot rectangular data as a color encoded matrix

temp_df = gap.pivot(index='country', columns='year', values='lifeExp')

# axes level function
plt.figure(figsize=(15, 10))
sns.heatmap(data=temp_df)
```

```python
# getting the heatmap only for a continent with annot
temp = gap[gap['continent']=='Europe'].pivot(index='country', columns='year', values='lifeExp')

plt.figure(figsize=(15, 10))
sns.heatmap(data=temp, annot=True, linewidth=0.5, cmap='summer')
```

```python
# clustermap
# plot the matrix dataset as a hierarchically-clustered heatmap
# this function requires scipy to be available
iris = px.data.iris()

sns.clustermap(iris.iloc[:,[0,1,2,3]]) # it shows the clustermap with dandograms

# the dandogram changes the order of the columns and rows on the basis of similarity,
# like, the closest columns of the data are petal length and sepal width
# and dis-similar are sepal length and petal width, it performs same as on the rows
# This clustering called Hierarchical Agglomerative Clustering
```

4. Regression Plot
    - Regplot
    - LMplot
    - Residplot

```python
# regplot
# It shows the relationship between two variables, it shows the linear relationship

sns.regplot(data=tips, x='total_bill', y='tip', order=2, color='magenta')
```

```python
# lmplot
# It fits a regression on the scatter plot
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
```

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
```

```python
# facet grid
sns.lmplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', col='sex', row='island', height=4)
```

```python
# Residplot
# it shows the residual after fitting the simple regression

sns.residplot(data=penguins, x='body_mass_g', y='bill_length_mm', lowess=True, line_kws=dict(color='r'))
```

6. Multi Plot Grids

    i. Facet Grids
        - FacetGrid

    ii. Pair Grids
        - Pairplot
        - PairGrid

    iii. Joind Grids
        - Jointplot
        - JointGrid

```python
# FacetGrid

fg= sns.FacetGrid(data=tips, col='sex', row='time')
fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
# Pairplot

sns.pairplot(data=penguins, hue='species') # kind can be used, hist, kde and also can be used marker ('o', 's', 'D') etc
```

```python
sns.pairplot(data=penguins, hue='species', corner=True, plot_kws=dict(marker='+'))
```

```python
# PairGrid
sns.PairGrid(data=penguins, hue='species').map(sns.histplot)
```

```python
# jointplot

sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='reg')
```

```python
# jointplot

sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
```

```python
# JointGrid

jg = sns.JointGrid(data=penguins, x='bill_length_mm', y='bill_depth_mm',hue='species')
jg.plot(sns.histplot, sns.boxplot)
```

```python
jg = sns.JointGrid(data=penguins, x='bill_length_mm', y='bill_depth_mm',hue='species')
jg.plot(sns.scatterplot, sns.histplot)
```

2. Categorical Plots

- Categorical Scatter Plot
    - Stripplot
    - Swarmplot
- Categorical Distribution Plots
    - Boxplot
    - Violinplot
- Categorical Estimate Plot -> for central tendency
    - Barplot
    - Pointplot
    - Countplot
- Figure level function -> catplot

```python
# Stripplot
# axes level function
sns.stripplot(data=tips, x='day', y='total_bill') # use, jitter=False to make the data into single line
```

```python
# stripplot
# figure level function
# adding jitter value
sns.catplot(data=tips, x='day', y='total_bill', kind='strip', jitter=.3, hue='sex')
```

```python
# swarmplot
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm', hue='sex')
```

```python
# axes level function
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex') # it mostly used for small dataset
```

```python
# Categorical Distribution Plots Boxplot Violinplot
'''A Boxplot is a standardized way of displaying the distribution of data
based on a five number summary (“minimum”, first quartile [Q1], median,
third quartile [Q3] and “maximum”). It can tell you about your outliers
and what their values are. Boxplots can also tell you if your data is symmetrical,
how tightly your data is grouped and if and how your data is skewed.'''

# Box plot
# axes level function
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
# figure level function
sns.catplot(data=tips, x='day', y='total_bill', kind='box', hue='sex')
```

```python
# single boxplot of numerical column

sns.boxplot(data=titanic, x='age')
```

```python
# Violinplot (Boxplot + KDEplot)
# axes level function
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
# figure level function
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='violin', split=True)
```

```python
# Categorical Estimate Plot -> for central tendency
# Barplot
# Pointplot
# Countplot
# Boxenplot
```

```python
# Barplot
import numpy as np
sns.barplot(data=tips, x='sex', y='total_bill') # errorbar can be removed
```

```python
sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker', estimator=np.std)
```

```python
sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker', ci=None) # removing the errorbar
```

```python
sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker')
```

```python
# pointplot
# sns.pointplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
sns.pointplot(data=tips, x='sex', y='total_bill', hue='smoker')
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
# countplot
'''A special case for the bar plot is when you want to show the number of
observations in each category rather than computing a statistic for a second variable.
This is similar to a histogram over a categorical, rather than quantitative, variable'''

sns.countplot(data=titanic, x='sex', hue='survived')
```

```python
sns.catplot(data=tips, x='sex', hue='day', kind='count')
```

```python
# boxenplot

sns.boxenplot(data=penguins, x='bill_length_mm', hue='sex')
```

```python
sns.boxenplot(data=tips, x='day', hue='sex')
```

```python
import pandas as pd
import plotly.express as px 
import plotly.graph_objects as go 
import numpy as np 
import os 
import seaborn as sns 
import matplotlib.pyplot as plt 
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = px.data.stocks()
df.head()
```

```python
# Line plot 
fig1 = px.line(df,
               x='date',
               y='GOOG',
               labels={
                   'date':'Date',
                   'GOOG':'Stock Price'
               }
               )
fig1.show()
```

```python
fig2 = px.line(
    df,
    x='date',
    y=['GOOG', 'AAPL'],
    labels={
        'date':'DATE',
        'value':'STOCKS',
    },
    title='GOOGLE vs APPLE STOCKS'
)

fig2.update_layout(
    xaxis=dict(
        showgrid=False,
        showticklabels=True,
        ticks='outside',
        tickfont=dict(
            family='Ariel',
            size=17,
            color='white'
        ),
        
        showline=True,
        linewidth=5,
        linecolor='rgb(192, 192, 192)'
    ),
    yaxis=dict(
        showgrid=False,
        showticklabels=True,
        ticks='outside',
        tickfont=dict(
            family='Ariel',
            size=17,
            color='white'
        ),

        showline=True,
        linewidth=5,
        linecolor='rgb(82, 82, 82)'
    ),

    plot_bgcolor='white',
    paper_bgcolor='white'
)
fig2.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df.date,
    y=df.NFLX,
    mode='lines',
    name='Netflix',
    ))

fig.add_trace(go.Scatter(
    x=df.date,
    y=df.GOOG,
    mode='lines+markers',
    name='Google',
))

fig.add_trace(go.Scatter(
    name='Amazon',
    line=dict(
        color='firebrick',
        width=2,
        dash='dashdot'
    )
))

fig.update_layout(
    xaxis=dict(
        showgrid=False,
        showticklabels=True,
        ticks='outside',
        tickfont=dict(
            family='Times_New_Roman',
            size=15,
            color='blue'
        ),
        showline=True,
        linewidth=5,
        linecolor='rgb(164, 164, 164)'
    ),
    yaxis=dict(
        showgrid=False,
        showticklabels=True,
        ticks='inside',
        tickfont=dict(
            family='calibiri',
            size=18,
            color='yellow'
        ),
        showline=True,
        linewidth=5,
        linecolor='rgb(182, 182, 182)'
    ),

    margin=dict(
        autoexpand=False,
        l=80,
        t=80,
        b=80,
        r=120
    ),
    plot_bgcolor='gray',
    paper_bgcolor='gray',
    showlegend=True,
)

fig.show()
```
